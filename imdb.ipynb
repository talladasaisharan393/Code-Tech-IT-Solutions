```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('IMDb_top_movies.csv')
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
lbec = LabelEncoder()

lbec.fit(df['movie title'])
movie_title_encoded = lbec.transform(df['movie title'])
movie_title_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
lbec = LabelEncoder()

lbec.fit(df['director'])
director_encoded = lbec.transform(df['director'])
director_encoded
```

```python
df['movie title'] = movie_title_encoded
df['director'] = director_encoded
df.head()
```

```python
df.corr()
```

```python
X = df[["movie title",'director']]
Y = df["imdb rating"]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,train_size = 0.80)
```

```python
model = LinearRegression()

model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

ytrainPred = model.predict(xtrain)
ytestPred = model.predict(xtest)


#Calculate Mean Squared Error for train & test data respectively
mse_train = mean_squared_error(ytrain,ytrainPred)
mse_test = mean_squared_error(ytest,ytestPred)

print("MeanSquaredError_Train" , mse_train , "<------>" , "MeanSquaredError_Test" , mse_test)

#Calculate R-Squared Value for train & test data respectively

r2_train = r2_score(ytrain,ytrainPred)
r2_test = r2_score(ytest,ytestPred)

print("R2Score_Train" , r2_train , "<------>" , "R2Score_Test" , r2_test)
```
